```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def csv_to_df(data_dir):
    """
    loading all csv files in a dir and converting them to df
    :return: df
    """
    df = pd.DataFrame()
    df_buffer = pd.DataFrame()
    # appending all csv files from one ds into one dfy
    for file in sorted(os.listdir(data_dir), reverse=True):
        if ".csv" in file:
            df_buffer = pd.read_csv(data_dir + file)
            df = pd.concat([df_buffer, df], axis=0)

    df.reset_index(drop=True, inplace=True)
    return df
```

# PT1-element

```python
df = csv_to_df("./data/pt1/")
cols = df.columns
df = df.rename(columns={cols[1]:"param_T", cols[2]:"y", cols[3]:"x"})

x1 = df["x"]
param_T1 = df["param_T"]
y1 = df["y"]



plt.figure(figsize=(20,5))
plt.subplot(3, 1, 1)
plt.subplots_adjust(hspace=0.5)
plt.plot(x1, label="input")
plt.legend(loc="upper right")
plt.title("input")

plt.subplot(3, 1, 2)
plt.plot(param_T1, label="param_T")
plt.legend(loc="upper right")
plt.title("parameters")

plt.subplot(3, 1, 3)
plt.plot(y1, label="output")
plt.legend(loc="upper right")
plt.title("output")
```

```python
# export to csv
df = pd.DataFrame()

df["x"] = x1
df["param_T"] = param_T1
df["y"] = y1

df.to_csv("./data/pt1_merged.csv", index=False)
```

# PT2-element

```python
df = csv_to_df("./data/pt2/")
cols = df.columns
df = df.rename(columns={cols[1]:"x", cols[2]:"y", cols[4]:"param_w", cols[3]:"param_D"})

x2 = df["x"]
param_w2 = df["param_w"]
param_D2 = df["param_D"]
y2 = df["y"]

plt.figure(figsize=(20,5))
plt.subplot(3, 1, 1)
plt.subplots_adjust(hspace=0.5)
plt.plot(x2, label="input")
plt.legend(loc="upper right")
plt.title("input")

plt.subplot(3, 1, 2)
plt.plot(param_w2, label="param_w")
plt.plot(param_D2, label="param_D")
plt.legend(loc="upper right")
plt.title("parameters")

plt.subplot(3, 1, 3)
plt.plot(y2, label="output")
plt.legend(loc="upper right")
plt.title("output")
```

```python
# export to csv
df = pd.DataFrame()

df["x"] = x2
df["param_w"] = param_w2
df["param_D"] = param_D2
df["y"] = y2

df.to_csv("./data/pt2_merged.csv", index=False)
```
